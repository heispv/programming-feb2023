```python
# Creating a Series
HP = [150, 400, 250]

HP_series = pd.Series(HP)
HP_series
```

```python
# Creating a Series and use a specified index
HP = [150, 400, 250]

new_HP_Series = pd.Series(HP, index=['first', 'second','third'])
new_HP_Series
```

```python
new_HP_Series['second']
```

```python
# Creating a Series using a dictionary
dict_HP = {
    'first.': 150,
    'second.': 400,
    'third.': 250
}

dict_HP_Series = pd.Series(dict_HP)
dict_HP_Series
```

```python
# Creating a Series using a part of a dictionary
dict_HP = {
    'first.': 150,
    'second.': 400,
    'third.': 250
}

dict_HP_Series = pd.Series(dict_HP, index=['first.', 'second.'])
dict_HP_Series
```
