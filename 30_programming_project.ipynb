## Changing the index

## Plotting the Candlestick Chart

```python
mpf.plot(data_example, type='candle', style='yahoo', title="Candlestick Chart")
mpf.show()
```

```python
data['is_green'] = data.apply(lambda row: 1 if row['close'] - row['open'] > 0 else 0, axis=1)
data[:20]
```

## Adding signals column

```python
signals = [''] * len(data)

for i in range(0, len(data)-2):
    if (data['is_green'][i] + data['is_green'][i+1] + data['is_green'][i+2]) == 3:
        signals[i+2] = 'buy'
    elif (data['is_green'][i] + data['is_green'][i+1] + data['is_green'][i+2]) == 0:
        signals[i+2] = 'sell'
    else:
        signals[i+2] = 'neutral'

data['buy_sell'] = signals
```

```python
data[:20]
```

```python
raw_data = data.copy()
```

## Adding TP and SL columns

```python
sl = [''] * len(data)
tp = [''] * len(data)

for i in range(2, len(data)):
    if data['buy_sell'][i] == 'sell':
        sl[i] = data['high'][i-2]
        tp_amount = (data['high'][i-2] - data['close'][i]) * 2
        tp[i] = data['close'][i] - tp_amount
    elif data['buy_sell'][i] == 'buy':
        sl[i] = data['low'][i-2]
        tp_amount = (data['close'][i] - data['low'][i-2]) * 2
        tp[i] = data['close'][i] + tp_amount


data['tp'] = tp
data['sl'] = sl
```

```python
data[:30]
```

# Backtesting the Strategy

## Adding PNL column

```python
pnl = [0] * len(data)

for i in range(len(data)-1000):
    row = data.iloc[i]
    if row['buy_sell'] in ['sell', 'buy']:
        tp = row['tp']
        sl = row['sl']
    else:
        continue

    j = i + 1
    while True:
        row_to_check = data.iloc[j]
        high = row_to_check['high']
        low = row_to_check['low']

        if low < tp < high:
            pnl[i] = 2
            break
        elif low < sl < high:
            pnl[i] = -1
            break
        else:
            j += 1

data['pnl'] = pnl
```

```python
data[:40]
```

```python
data['pnl']
```

```python
sum = [0]
for i in range(len(data)-500):
    sum.append(sum[i] + data['pnl'][i])
```

## Plot the PNL plot

```python
plt.plot(sum)
plt.title('3 Consecutve Candles Strategy')
plt.xlabel('Total PNL')
plt.ylabel('# of Candle')
plt.show()
```
